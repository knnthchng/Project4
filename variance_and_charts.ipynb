```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('Resources/train.csv')
data
```

```python
# Rename columns for better comprehensibility
data = data.rename(columns={"Designation": "Rank",
                           "Resource Allocation":"Hours Worked Per Day",
                           "Burn Rate": "Burnout Rate"})
```

```python
# Mean of all numerical variables

values_rank = data['Rank']
values_hours = data['Hours Worked Per Day']
values_fatigue = data['Mental Fatigue Score']
values_burnout = data['Burnout Rate']

mean_values = np.mean(values_rank)
mean_hours = np.mean(values_hours)
mean_fatigue = np.mean(values_fatigue)
mean_burnout = np.mean(values_burnout)

print("Rank Average:", mean_values)
print("Average Hours Worked per Day:", mean_hours)
print("Mental Fatigue Score Average:", mean_fatigue)
print("Burnout Rate Average:", mean_burnout)
```

```python
# Median of all numerical values
median_rank = values_rank.median()
median_hours = values_hours.median()
median_fatigue = values_fatigue.median()
median_burnout = values_burnout.median()

print("Designation Median:", median_rank)
print("Resource Allocation Median:", median_hours)
print("Mental Fatigue Score Median:", median_fatigue)
print("Burn Rate Median:", median_burnout)
```

```python
# Mode of all variables
mode_value = data.mode().iloc[0]

print("Mode:", mode_value)
```

```python
# Standard Deviation of numerical variables

std_rank = np.std(values_rank)
std_hours = np.std(values_hours)
std_fatigue = np.std(values_fatigue)
std_burnout = np.std(values_burnout)

print("Designation Standard Deviation:", std_rank)
print("Resource Allocation Standard Deviation:", std_hours)
print("Mental Fatigue Score Standard Deviation:", std_fatigue)
print("Burn Rate Standard Deviation:", std_burnout)
```

```python
# Gender Composition

column_values = data['Gender']

value_counts = column_values.value_counts()

plt.figure(figsize=(8, 6))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Gender')
plt.show()

save_path = 'Visuals/fig_1.png'
plt.savefig(save_path, dpi=300, bbox_inches='tight')
```
